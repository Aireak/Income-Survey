```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
expense_data=pd.read_csv('survey.responses.csv')
```

```python
expense_data.head()
```

```python
expense_data.describe()
```

```python
# Show the ages with the highest income
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='income', data=expense_data)
plt.title('Income by Age')
plt.show()
plt.savefig('ages_income_distribution.png')

# Show the gender distribution across spending categories
expense_columns = ['expenses.utilities', 'expenses.entertainment', 'expenses.school_fees', 'expenses.shopping', 'expenses.healthcare']
df_expense = expense_data.melt(id_vars=['gender'], value_vars=expense_columns, var_name='category', value_name='amount')

plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='amount', hue='gender', data=df_expense)
plt.title('Spending Distribution by Gender')
plt.show()
plt.savefig('gender_distribution_across_spending.png')


```

```python
# The chart shows that ages 55 and 33 have the highest distribution
## The second chart shows that women spend more in entertainment and shopping while men spend more for utilities and healthcare
```
